```python
#write a program to find a palindrome of a number
def is_palindrome(num):
    original_num = num
    reversed_num = 0

    while num > 0:
        digit = num % 10
        reversed_num = reversed_num * 10 + digit
        num //= 10

    return original_num == reversed_num

number = int(input("Enter a number to check if it's a palindrome: "))
if is_palindrome(number):
    print(f"{number} is a palindrome.")
else:
    print(f"{number} is not a palindrome.")




```

```python
#program to check the number is armstrong or not
def is_armstrong(num):
    digits = [int(d) for d in str(num)]
    power = len(digits)
    total = sum(d ** power for d in digits)
    return total == num

# Input from user
number = int(input("Enter a number to check if it's an Armstrong number: "))

# Check and display result
if is_armstrong(number):
    print(f"{number} is an Armstrong number.")
else:
    print(f"{number} is not an Armstrong number.")
```

```python
#30/07/25 WATER JUG PROBLEM
def get_int_input(prompt, min_val=0):
    while True:
        try:
            val = int(input(prompt))
            if val < min_val:
                print(f"Please enter a value >= {min_val}.")
            else:
                return val
        except ValueError:
            print("Please enter a valid integer.")

def print_menu():
    print("\nList of Operations:")
    print("1. Fill Jug A completely")
    print("2. Fill Jug B completely")
    print("3. Empty Jug A completely")
    print("4. Empty Jug B completely")
    print("5. Pour from Jug A to Jug B (until B is full or A is empty)")
    print("6. Pour from Jug B to Jug A (until A is full or B is empty)")
    print("7. Pour all from Jug B to Jug A")
    print("8. Pour all from Jug A to Jug B")
    print("9. Exit")

# Initial setup with input validation
a = get_int_input("Enter Jug A Capacity: ")
b = get_int_input("Enter Jug B Capacity: ")

ai = get_int_input(f"Initially Water in Jug A (0 to {a}): ", 0)
while ai > a:
    ai = get_int_input(f"Value cannot exceed Jug A capacity ({a}): ")

bi = get_int_input(f"Initially Water in Jug B (0 to {b}): ", 0)
while bi > b:
    bi = get_int_input(f"Value cannot exceed Jug B capacity ({b}): ")

af = get_int_input(f"Final State of Jug A (0 to {a}): ", 0)
while af > a:
    af = get_int_input(f"Value cannot exceed Jug A capacity ({a}): ")

bf = get_int_input(f"Final State of Jug B (0 to {b}): ", 0)
while bf > b:
    bf = get_int_input(f"Value cannot exceed Jug B capacity ({b}): ")

steps = 0

# Process operations
while ai != af or bi != bf:
    print(f"\nCurrent State: Jug A = {ai}, Jug B = {bi}")
    print_menu()
    op = get_int_input("Enter the operation (1 - 9): ")

    if op == 1:
        ai = a
    elif op == 2:
        bi = b
    elif op == 3:
        ai = 0
    elif op == 4:
        bi = 0
    elif op == 5:
        transfer = min(ai, b - bi)
        ai -= transfer
        bi += transfer
    elif op == 6:
        transfer = min(bi, a - ai)
        bi -= transfer
        ai += transfer
    elif op == 7:
        if ai + bi > a:
            print("Cannot pour all from Jug B to Jug A (overflows).")
        else:
            ai += bi
            bi = 0
    elif op == 8:
        if ai + bi > b:
            print("Cannot pour all from Jug A to Jug B (overflows).")
        else:
            bi += ai
            ai = 0
    elif op == 9:
        break
    else:
        print("Invalid operation. Choose from 1 to 9.")
        continue

    steps += 1
    print(f"After operation {steps}: Jug A = {ai}, Jug B = {bi}")

# Final state confirmation
if ai == af and bi == bf:
    print(f"\nGoal achieved in {steps} steps! Final State: Jug A = {ai}, Jug B = {bi}")
```

```python
#13-08-25 river crossing puzzle with sheep,wolf and cabbage
from collections import deque
def is_valid(state):
    f, w, g, c = state
    # If farmer is not with goat
    if f != g:
        if g == w:
            return False  # Wolf eats goat
        if g == c:
            return False  # Goat eats cabbage
    return True

def get_next_states(state):
    f, w, g, c = state
    positions = [f, w, g, c]
    possible_moves = []

    def toggle(side):  # move from L to R or R to L
        return 'R' if side == 'L' else 'L'

    # Farmer moves alone
    new_state = (toggle(f), w, g, c)
    if is_valid(new_state):
        possible_moves.append(new_state)

    # Farmer takes wolf
    if f == w:
        new_state = (toggle(f), toggle(w), g, c)
        if is_valid(new_state):
            possible_moves.append(new_state)

    # Farmer takes goat
    if f == g:
        new_state = (toggle(f), w, toggle(g), c)
        if is_valid(new_state):
            possible_moves.append(new_state)

    # Farmer takes cabbage
    if f == c:
        new_state = (toggle(f), w, g, toggle(c))
        if is_valid(new_state):
            possible_moves.append(new_state)

    return possible_moves

def bfs(start, goal):
    queue = deque()
    queue.append((start, [start]))
    visited = set()

    while queue:
        current, path = queue.popleft()
        if current == goal:
            return path
        visited.add(current)
        for next_state in get_next_states(current):
            if next_state not in visited:
                queue.append((next_state, path + [next_state]))

    return None

# Run the puzzle
start_state = ('L', 'L', 'L', 'L')
goal_state = ('R', 'R', 'R', 'R')
solution = bfs(start_state, goal_state)

# Display solution
if solution:
    print("Solution path:")
    for step in solution:
        print(step)
else:
    print("No solution found.")

```

```python
from collections import deque

# State representation: (farmer_side, wolf_side, goat_side, cabbage_side)
# 0 for left bank, 1 for right bank

def is_valid_state(state):
    farmer_side, wolf_side, goat_side, cabbage_side = state

    # Check if wolf eats goat on either side
    if wolf_side == goat_side and wolf_side != farmer_side:
        return False
    # Check if goat eats cabbage on either side
    if goat_side == cabbage_side and goat_side != farmer_side:
        return False
    return True

def solve_river_crossing():
    initial_state = (0, 0, 0, 0)  # All on left bank
    goal_state = (1, 1, 1, 1)    # All on right bank

    queue = deque([(initial_state, [])])  # (current_state, path_to_reach_state)
    visited = {initial_state}

    while queue:
        current_state, path = queue.popleft()

        if current_state == goal_state:
            return path + [current_state]

        farmer_side = current_state[0]
        other_side = 1 - farmer_side # The opposite bank

        # Possible moves: (item_to_move, new_state_tuple_index)
        moves = [
            (None, None),  # Farmer crosses alone
            (1, 1),        # Farmer and Wolf
            (2, 2),        # Farmer and Goat
            (3, 3)         # Farmer and Cabbage
        ]

        for item_index, state_index in moves:
            next_state_list = list(current_state)
            next_state_list[0] = other_side # Move farmer

            if item_index is not None:
                # Check if item is on the same side as the farmer to move it
                if current_state[item_index] == farmer_side:
                    next_state_list[item_index] = other_side
                else:
                    continue # Cannot move this item

            next_state = tuple(next_state_list)

            if is_valid_state(next_state) and next_state not in visited:
                visited.add(next_state)
                queue.append((next_state, path + [current_state]))
    return None # No solution found

# Example usage:
solution = solve_river_crossing()
if solution:
    print("Solution found:")
    for i, state in enumerate(solution):
        print(f"Move {i}: Farmer: {'Right' if state[0] else 'Left'}, Wolf: {'Right' if state[1] else 'Left'}, Goat: {'Right' if state[2] else 'Left'}, Cabbage: {'Right' if state[3] else 'Left'}")
else:
    print("No solution exists.")
```

```python
class Node:
    def __init__(self, data, level, fval):
        self.data = data
        self.level = level
        self.fval = fval

    def generate_child(self):
        x, y = self.find(self.data, '_')
        val_list = [[x, y - 1], [x, y + 1], [x - 1, y], [x + 1, y]]
        children = []

        for i in val_list:
            child = self.shuffle(self.data, x, y, i[0], i[1])
            if child is not None:
                child_node = Node(child, self.level + 1, 0)
                children.append(child_node)
        return children

    def shuffle(self, puz, x1, y1, x2, y2):
        if 0 <= x2 < len(puz) and 0 <= y2 < len(puz[0]):
            temp_puz = self.copy(puz)
            temp_puz[x1][y1], temp_puz[x2][y2] = temp_puz[x2][y2], temp_puz[x1][y1]
            return temp_puz
        return None

    def copy(self, root):
        return [row[:] for row in root]

    def find(self, puz, x):
        for i in range(len(puz)):
            for j in range(len(puz[0])):
                if puz[i][j] == x:
                    return i, j


class Puzzle:
    def __init__(self, size):
        self.n = size
        self.open = []
        self.closed = []

    def accept(self):
        puz = []
        print(f"Enter {self.n} rows of the puzzle, each with {self.n} characters (use '_' for the blank space):")
        for i in range(self.n):
            while True:
                temp_str = input(f"Row {i + 1}: ")
                if len(temp_str) != self.n:
                    print(f"Invalid input. Please enter exactly {self.n} characters.")
                else:
                    puz.append(list(temp_str))
                    break
        return puz

    def f(self, start, goal):
        return self.h(start.data, goal) + start.level

    def h(self, start, goal):
        temp = 0
        for i in range(self.n):
            for j in range(self.n):
                if start[i][j] != goal[i][j] and start[i][j] != '_':
                    temp += 1
        return temp

    def process(self):
        print("Enter the start state matrix:")
        start = self.accept()
        print("Enter the goal state matrix:")
        goal = self.accept()

        start_node = Node(start, 0, 0)
        start_node.fval = self.f(start_node, goal)
        self.open.append(start_node)

        print("\nSolving...\n")
        while True:
            cur = self.open[0]
            print("Step (f =", cur.fval, "):")
            for row in cur.data:
                print(" ".join(row))
            print("")

            if self.h(cur.data, goal) == 0:
                print("Goal state reached!")
                break

            for child in cur.generate_child():
                child.fval = self.f(child, goal)
                self.open.append(child)

            self.closed.append(cur)
            self.open.pop(0)
            self.open.sort(key=lambda x: x.fval)


# Run the puzzle
puz = Puzzle(3)
puz.process()
```

```python
from collections import deque

def is_valid(state):
    f, w, g, c = state
    if f != g:
        if g == w or g == c:
            return False
    return True

def toggle(side):
    return 'R' if side == 'L' else 'L'

def get_next_states(state):
    f, w, g, c = state
    possible_moves = []

    # Farmer moves alone
    new_state = (toggle(f), w, g, c)
    if is_valid(new_state):
        possible_moves.append(new_state)

    # Farmer takes wolf
    if f == w:
        new_state = (toggle(f), toggle(w), g, c)
        if is_valid(new_state):
            possible_moves.append(new_state)

    # Farmer takes goat
    if f == g:
        new_state = (toggle(f), w, toggle(g), c)
        if is_valid(new_state):
            possible_moves.append(new_state)

    # Farmer takes cabbage
    if f == c:
        new_state = (toggle(f), w, g, toggle(c))
        if is_valid(new_state):
            possible_moves.append(new_state)

    return possible_moves

def dfs(current, goal, path, visited):
    if current == goal:
        return path
    visited.add(current)
    for next_state in get_next_states(current):
        if next_state not in visited:
            result = dfs(next_state, goal, path + [next_state], visited)
            if result:
                return result
    return None

# Run the puzzle using DFS
start_state = ('L', 'L', 'L', 'L')
goal_state = ('R', 'R', 'R', 'R')
solution = dfs(start_state, goal_state, [start_state], set())

# Display solution
if solution:
    print("Solution path:")
    for step in solution:
        print(step)
else:
    print("No solution found.")
```

```python
import math
# Initialize board
def create_board():
    return [[" " for _ in range(3)] for _ in range(3)]

# Print board
def print_board(board):
    for row in board:
        print("|".join(row))
        print("-" * 5)

# Check winner
def check_winner(board):
    # Rows & Columns
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != " ":
            return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] != " ":
            return board[0][i]
    # Diagonals
    if board[0][0] == board[1][1] == board[2][2] != " ":
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != " ":
        return board[0][2]
    return None

# Check if moves left
def is_moves_left(board):
    return any(" " in row for row in board)

# Minimax function
def minimax(board, depth, is_maximizing):
    winner = check_winner(board)
    if winner == "X":
        return 1
    if winner == "O":
        return -1
    if not is_moves_left(board):
        return 0

    if is_maximizing:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == " ":
                    board[i][j] = "X"
                    best = max(best, minimax(board, depth + 1, False))
                    board[i][j] = " "
        return best
    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == " ":
                    board[i][j] = "O"
                    best = min(best, minimax(board, depth + 1, True))
                    board[i][j] = " "
        return best

# Best move for AI
def find_best_move(board):
    best_val = -math.inf
    best_move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if board[i][j] == " ":
                board[i][j] = "X"
                move_val = minimax(board, 0, False)
                board[i][j] = " "
                if move_val > best_val:
                    best_move = (i, j)
                    best_val = move_val
    return best_move

# Main game loop
def play_game():
    board = create_board()
    human = "O"
    ai = "X"

    print("Welcome to Tic-Tac-Toe with Minimax AI!")
    print_board(board)

    while True:
        # Human move
        row = int(input("Enter row (0-2): "))
        col = int(input("Enter col (0-2): "))
        if board[row][col] == " ":
            board[row][col] = human
        else:
            print("Cell occupied! Try again.")
            continue

        print_board(board)
        if check_winner(board):
            print("You win! üéâ")
            break
        if not is_moves_left(board):
            print("It's a draw!")
            break

        # AI move
        print("AI is making a move...")
        move = find_best_move(board)
        board[move[0]][move[1]] = ai

        print_board(board)
        if check_winner(board):
            print("AI wins! ü§ñ")
            break
        if not is_moves_left(board):
            print("It's a draw!")
            break

# Run the game
if __name__ == "__main__":
    play_game()
```

```python
#tic-tac-toe game using alpha-beta pruning
import math

def create_board():
    return [[" " for _ in range(3)] for _ in range(3)]

def print_board(board):
    for row in board:
        print("|".join(row))
        print("-" * 5)

def check_winner(board):
    # Rows & Columns
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != " ":
            return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] != " ":
            return board[0][i]
    # Diagonals
    if board[0][0] == board[1][1] == board[2][2] != " ":
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != " ":
        return board[0][2]
    return None

def is_moves_left(board):
    return any(" " in row for row in board)

def minimax_alpha_beta(board, depth, alpha, beta, is_maximizing):
    winner = check_winner(board)
    if winner == "X":
        return 1
    if winner == "O":
        return -1
    if not is_moves_left(board):
        return 0

    if is_maximizing:
        max_eval = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == " ":
                    board[i][j] = "X"
                    eval = minimax_alpha_beta(board, depth + 1, alpha, beta, False)
                    board[i][j] = " "
                    max_eval = max(max_eval, eval)
                    alpha = max(alpha, eval)
                    if beta <= alpha:
                        break
            if beta <= alpha:
                break
        return max_eval
    else:
        min_eval = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == " ":
                    board[i][j] = "O"
                    eval = minimax_alpha_beta(board, depth + 1, alpha, beta, True)
                    board[i][j] = " "
                    min_eval = min(min_eval, eval)
                    beta = min(beta, eval)
                    if beta <= alpha:
                        break
            if beta <= alpha:
                break
        return min_eval

def find_best_move_alpha_beta(board):
    best_val = -math.inf
    best_move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if board[i][j] == " ":
                board[i][j] = "X"
                move_val = minimax_alpha_beta(board, 0, -math.inf, math.inf, False)
                board[i][j] = " "
                if move_val > best_val:
                    best_move = (i, j)
                    best_val = move_val
    return best_move

def play_game_alpha_beta():
    board = create_board()
    human = "O"
    ai = "X"

    print("Welcome to Tic-Tac-Toe with Alpha-Beta Pruning AI!")
    print_board(board)

    while True:
        # Human move
        row = int(input("Enter row (0-2): "))
        col = int(input("Enter col (0-2): "))
        if board[row][col] == " ":
            board[row][col] = human
        else:
            print("Cell occupied! Try again.")
            continue

        print_board(board)
        if check_winner(board):
            print("You win! üéâ")
            break
        if not is_moves_left(board):
            print("It's a draw!")
            break

        # AI move
        print("AI is making a move...")
        move = find_best_move_alpha_beta(board)
        board[move[0]][move[1]] = ai

        print_board(board)
        if check_winner(board):
            print("AI wins! ü§ñ")
            break
        if not is_moves_left(board):
            print("It's a draw!")
            break

# Run the game
if __name__ == "__main__":
    play_game_alpha_beta()
```

```python
import math

def check_winner(board):
    win_states = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],
        [0, 3, 6], [1, 4, 7], [2, 5, 8],
        [0, 4, 8], [2, 4, 6]
    ]
    for (x, y, z) in win_states:
        if board[x] == board[y] == board[z] and board[x] != ' ':
            return board[x]
    if ' ' not in board:
        return 'Draw'
    return None

def minimax(board, depth, alpha, beta, is_maximizing):
    winner = check_winner(board)
    if winner == 'X':
        return -10 + depth
    elif winner == 'O':
        return 10 - depth
    elif winner == 'Draw':
        return 0
    if is_maximizing:
        max_eval = -math.inf
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'O'
                eval = minimax(board, depth + 1, alpha, beta, False)
                board[i] = ' '
                max_eval = max(max_eval, eval)
                alpha = max(alpha, eval)
                if beta <= alpha:
                    break
        return max_eval
    else:
        min_eval = math.inf
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'X'
                eval = minimax(board, depth + 1, alpha, beta, True)
                board[i] = ' '
                min_eval = min(min_eval, eval)
                beta = min(beta, eval)
                if beta <= alpha:
                    break
        return min_eval

def best_move(board):
    best_val = -math.inf
    move = None
    for i in range(9):
        if board[i] == ' ':
            board[i] = 'O'
            move_val = minimax(board, 0, -math.inf, math.inf, False)
            board[i] = ' '
            if move_val > best_val:
                best_val = move_val
                move = i
    return move

def print_board (board):
    for i in range(0, 9, 3):
        print(board[i], '|', board[i+1], '|', board[i+2])
        if i < 6:
            print('--+---+--')

def play_game():
    board = [' '] * 9
    while True:
        print_board(board)
        try:
            move = int(input("Enter your move (1-9): ")) - 1
            if move < 0 or move > 8 or board[move] != ' ':
                print("Invalid move. Try again.")
                continue
        except ValueError:
            print("Please enter a number between 1 and 9.")
            continue
        board[move] = 'X'
        result = check_winner(board)
        if result:
            print_board(board)
            print("Result:", result)
            break
        ai_move = best_move(board)
        board[ai_move] = 'O'
        result = check_winner(board)
        if result:
            print_board(board)
            print("Result:", result)
            break

play_game()
```

```python
class MapColoringCSP:
    def __init__(self, variables, domains, neighbors):
        # Variables: regions to color
        self.variables = variables

        # Domains: possible colors for each region
        self.domains = domains

        # Neighbors: adjacency list for each region
        self.neighbors = neighbors

        # Assignment: keeps track of the color assigned to each region
        self.assignment = {}

    def is_consistent(self, variable, color):
        """Check if the color assignment is consistent with the neighbors."""
        for neighbor in self.neighbors[variable]:
            # Check if the neighbor is already assigned the same color
            if neighbor in self.assignment and self.assignment[neighbor] == color:
                return False
        return True

    def backtrack(self):
        """Backtracking search to assign colors to regions."""
        # If every variable has been assigned a color, return the solution
        if len(self.assignment) == len(self.variables):
            return self.assignment

        # Select an unassigned variable (region)
        variable = self.select_unassigned_variable()

        # Try each color in the domain
        for color in self.domains[variable]:
            if self.is_consistent(variable, color):
                # Assign the color if it's consistent
                self.assignment[variable] = color
                # Recursively attempt to assign colors to the remaining variables
                result = self.backtrack()
                if result:
                    return result
                # If no solution, remove the current assignment and backtrack
                del self.assignment[variable]
        return None  # No solution found

    def select_unassigned_variable(self):
        """Select the next unassigned variable."""
        for variable in self.variables:
            if variable not in self.assignment:
                return variable

    def solve(self):
        """Solve the CSP using backtracking."""
        return self.backtrack()


# Example: Coloring a simple map of Australia
def main():
    # Define the regions of the map (variables)
    variables = ['WA', 'NT', 'SA', 'Q', 'NSW', 'V', 'T']

    # Define the domains: possible colors for each region
    domains = {
        'WA': ['Red', 'Green', 'Blue'],
        'NT': ['Red', 'Green', 'Blue'],
        'SA': ['Red', 'Green', 'Blue'],
        'Q': ['Red', 'Green', 'Blue'],
        'NSW': ['Red', 'Green', 'Blue'],
        'V': ['Red', 'Green', 'Blue'],
        'T': ['Red', 'Green', 'Blue']
    }

    # Define neighbors: adjacent regions
    neighbors = {
        'WA': ['NT', 'SA'],
        'NT': ['WA', 'SA', 'Q'],
        'SA': ['WA', 'NT', 'Q', 'NSW', 'V'],
        'Q': ['NT', 'SA', 'NSW'],
        'NSW': ['Q', 'SA', 'V'],
        'V': ['SA', 'NSW'],
        'T': []  # Tasmania is isolated
    }

    # Initialize the CSP
    csp = MapColoringCSP(variables, domains, neighbors)

    # Solve the CSP
    solution = csp.solve()

    # Print the solution
    if solution:
        print("Solution found:")
        for region, color in solution.items():
            print(f"{region}: {color}")
    else:
        print("No solution found.")


if __name__ == "__main__":
    main()
```
