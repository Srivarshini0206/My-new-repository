```python
def temperature_modeling(a, b, c, time):
    # Calculate temperature based on time using the quadratic equation
    temperature = a * time**2 + b * time + c
    return temperature



a_hardcoded, b_hardcoded, c_hardcoded = 0.1, 2, 10

# Display results
print(" Hard-coded Variables for Weather Modeling")
time_hardcoded = 5  # Example time value
print("Temperature for hardcoded coefficients at time", time_hardcoded, "hours:", temperature_modeling(a_hardcoded, b_hardcoded, c_hardcoded, time_hardcoded))
print("\n")

print(" Keyboard Input for Weather Modeling")

a_keyboard = float(input("Enter coefficient a: "))
b_keyboard = float(input("Enter coefficient b: "))
c_keyboard = float(input("Enter coefficient c: "))

# Get time from user input
time_keyboard = float(input("Enter time in hours: "))

# Display results
print("Temperature for keyboard input coefficients at time", time_keyboard, "hours:", temperature_modeling(a_keyboard, b_keyboard, c_keyboard, time_keyboard))
print("\n")


# Create the weather.txt file with example data
with open('weather.txt', 'w') as file:
    file.write('0.5 3.0 15.0 7.0')

with open('weather.txt', 'r') as file:
    lines = file.readlines()
    a_file, b_file, c_file, time_file = map(float, lines[0].split())

# Display results
print("Read from a File for Weather Modeling")
print("Temperature for file input coefficients at time", time_file, "hours:", temperature_modeling(a_file, b_file, c_file, time_file))
print("\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def temperature_model(t, a, b, c):
    return a * t**2 + b * t + c
a = 0.5
b = 2
c = 10
time = 5
predicted_temperature = temperature_model(time, a, b, c)
print("Predicted Temperature:", predicted_temperature)
t_values = np.linspace(0, 10, 50)
temp_values = temperature_model(t_values, a, b, c)
plt.plot(t_values, temp_values)
plt.xlabel("Time")
plt.ylabel("Temperature")
plt.title("Weather Temperature Prediction using Quadratic Model")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
stages = ["Requirements", "Design", "Implementation", "Verification", "Maintenance"]
x_positions = np.linspace(-2, 2, len(stages))

def quadratic(x):
    return -x**2 + 10
y_positions = quadratic(x_positions)
x_curve = np.linspace(-2.5, 2.5, 400)
y_curve = quadratic(x_curve)
plt.figure(figsize=(10, 6))
plt.plot(x_curve, y_curve, linestyle="--", label="Quadratic Curve")
for i in range(len(stages)):
    if i > 0:

        plt.plot(
            [x_positions[i-1], x_positions[i]],
            [y_positions[i-1], y_positions[i-1]],
            linewidth=2
        )

        plt.plot(
            [x_positions[i], x_positions[i]],
            [y_positions[i-1], y_positions[i]],
            linewidth=2
        )

plt.scatter(x_positions, y_positions, zorder=5)
for i, stage in enumerate(stages):
    plt.text(x_positions[i], y_positions[i] + 0.3, stage, ha='center')
plt.title("Waterfall Model Representation Using Quadratic Curve")
plt.xlabel("Stages")
plt.ylabel("Value")
plt.grid(True)
plt.legend()
plt.show()
```
