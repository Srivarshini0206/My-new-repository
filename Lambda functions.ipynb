```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

def isStrong(n):
    original_num = n
    sum_of_factorials = 0
    while n > 0:
        digit = n % 10
        sum_of_factorials += factorial(digit)
        n //= 10
    return sum_of_factorials == original_num

def main():
    num = int(input("Enter a number: "))
    if isStrong(num):
        print(f"{num} is a strong number")
    else:
        print(f"{num} is not a strong number")

# Test cases
test_cases = []
for num in test_cases:
    if isStrong(num):
        print(f"{num} is a strong number")
    else:
        print(f"{num} is not a strong number")
```

```python
# Program to display the Fibonacci sequence up to n-th term

nterms = int(input("How many terms? "))

# first two terms
n1, n2 = 0, 1
count = 0

# check if the number of terms is valid
if nterms <= 0:
   print("Please enter a positive integer")
# if there is only one term, return n1
elif nterms == 1:
   print("Fibonacci sequence upto",nterms,":")
   print(n1)
# generate fibonacci sequence
else:
   print("Fibonacci sequence:")
   while count < nterms:
       print(n1)
       nth = n1 + n2
       # update values
       n1 = n2
       n2 = nth
       count += 1
```

```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta


start_date_str = input("Enter start date (YYYY-MM-DD): ")
end_date_str = input("Enter end date (YYYY-MM-DD): ")
start_date = datetime.strptime(start_date_str, "%Y-%m-%d")
end_date = datetime.strptime(end_date_str, "%Y-%m-%d")
num_days = (end_date - start_date).days + 1


dates = np.array([start_date + timedelta(days=i) for i in range(num_days)], dtype='datetime64[D]')
likes = np.zeros(num_days)
comments = np.zeros(num_days)
shares = np.zeros(num_days)


for i, date in enumerate(dates):
    print(f"Enter data for {date}:")
    likes[i] = int(input("Likes: "))
    comments[i] = int(input("Comments: "))
    shares[i] = int(input("Shares: "))


plt.figure(figsize=(10, 6))
plt.plot(dates, likes, label='Likes')
plt.plot(dates, comments, label='Comments')
plt.plot(dates, shares, label='Shares')
plt.title('Social Media Analytics Dashboard')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```
